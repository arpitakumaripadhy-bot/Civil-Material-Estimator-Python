```python
# Cell 1: Material Rate Database
# Rates are based on standard Indian market averages (INR)

material_rates = {
    "cement_bag": 420,          # per bag (50kg)
    "sand": 1500,               # per cubic meter (cum)
    "aggregate": 1800,          # per cubic meter (cum)
    "bricks": 8,                # per piece
    "steel_kg": 65              # per kg
}

print("✅ Material rates initialized successfully!")
print("Current Cement Rate: ₹",{material_rates['cement_bag']},"/(50kg)bag")
print("Current Sand Rate: ₹",{material_rates['sand']},"/cum")
print("Current aggregate Rate: ₹",{material_rates['aggregate']},"/cum")
print("Current Brick Rate: ₹",{material_rates['bricks']},"/piece")
print("Current Steel Rate: ₹",{material_rates['steel_kg']},"/kg")
```

```python
# Cell 2: Concrete Estimation Function

def estimate_concrete(length, width, thickness):
    # 1. Calculate Wet Volume
    wet_volume = length * width * thickness
    
    # 2. Convert to Dry Volume (Standard Civil Engineering Factor is 1.54)
    dry_volume = wet_volume * 1.54
    
    # 3. M20 Mix Ratio: 1 (Cement) + 1.5 (Sand) + 3 (Aggregate) = 5.5 total parts
    total_parts = 1 + 1.5 + 3
    
    # 4. Calculate Volumes of individual materials
    cement_volume = (1 / total_parts) * dry_volume
    sand_volume = (1.5 / total_parts) * dry_volume
    aggregate_volume = (3 / total_parts) * dry_volume
    
    # 5. Convert Cement Volume to Bags (1 bag of cement = 0.035 cubic meters)
    cement_bags = round(cement_volume / 0.035, 2)
    
    # 6. Round off other materials for ordering
    sand_needed = round(sand_volume, 2)
    aggregate_needed = round(aggregate_volume, 2)
    
    # 7. Calculate Costs using our material_rates dictionary
    cement_cost = cement_bags * material_rates["cement_bag"]
    sand_cost = sand_needed * material_rates["sand"]
    aggregate_cost = aggregate_needed * material_rates["aggregate"]
    total_cost = cement_cost + sand_cost + aggregate_cost
    
    # Return results as a dictionary
    return {
        "Volume (Cum)": round(wet_volume, 2),
        "Cement (Bags)": cement_bags,
        "Sand (Cum)": sand_needed,
        "Aggregate (Cum)": aggregate_needed,
        "Total Cost (INR)": round(total_cost, 2)
    }

print("✅ Concrete estimation function is ready!")
```

```python
# Cell 3: Main Program User Interface

print("--- Welcome to CivicEstimator v1.0 ---")
print("Please enter the dimensions of the Concrete Slab in meters.")

# Taking User Input and applying Type Conversion (String to Float)
l = float(input("Enter Length (m): "))
w = float(input("Enter Width (m): "))
t = float(input("Enter Thickness/Depth (m): "))

# Call our function and store results
project_report = estimate_concrete(l, w, t)

print("\n" + "="*40)
print("       BILL OF QUANTITIES (BOQ)       ")
print("="*40)

# Using a For Loop to dynamically print out our report items
for material, quantity in project_report.items():
    if material == "Total Cost (INR)":
        continue # We will print total cost at the very end
    print(f" {material:<20} : {quantity}")

print("-"*40)
print(f" ESTIMATED BUDGET TOTAL: ₹{project_report['Total Cost (INR)']}/-")
print("="*40)
print("\n*Note: Calculation includes a 1.54 dry volume shrinkage factor.")
```

```python
# Cell 4: Brick Wall Estimation Function

def estimate_brickwork(length, height, thickness):
    # 1. Calculate the total volume of the wall
    wall_volume = length * height * thickness
    
    # 2. Standard civil engineering rule: 500 bricks per cubic meter
    total_bricks = round(wall_volume * 500)
    
    # 3. Mortar volume is roughly 30% of total wall volume
    mortar_volume = wall_volume * 0.30
    dry_mortar_volume = mortar_volume * 1.33 # 1.33 is the dry volume factor for mortar
    
    # 4. Standard Mortar Mix 1:3 (1 Part Cement, 3 Parts Sand) -> Total 4 parts
    total_parts = 1 + 3
    cement_volume = (1 / total_parts) * dry_mortar_volume
    sand_volume = (3 / total_parts) * dry_mortar_volume
    
    # 5. Convert Cement Volume to Bags (1 bag = 0.035 cum)
    cement_bags = round(cement_volume / 0.035, 2)
    sand_needed = round(sand_volume, 2)
    
    # 6. Calculate Costs using our material_rates dictionary
    brick_cost = total_bricks * material_rates["bricks"]
    cement_cost = cement_bags * material_rates["cement_bag"]
    sand_cost = sand_needed * material_rates["sand"]
    total_cost = brick_cost + cement_cost + sand_cost
    
    return {
        "Wall Volume (Cum)": round(wall_volume, 2),
        "Total Bricks (Nos)": total_bricks,
        "Cement (Bags)": cement_bags,
        "Sand (Cum)": sand_needed,
        "Total Cost (INR)": round(total_cost, 2)
    }

print("✅ Brickwork estimation function is ready!")
```

```python
### Cell 5: Main Application Interface with Flow Control

print("=========================================")
print("     WELCOME TO CIVICESTIMATOR v2.0     ")
print("=========================================")
print("What would you like to estimate today?")
print("1. Concrete Slab / Footing")
print("2. Brick Wall")
print("=========================================")

# Get user choice
choice = input("Enter choice (1 or 2): ")

# Setup a blank dictionary to hold our final report
project_report = {}

# --- FLOW CONTROL BLOCK ---
if choice == "1":
    print("\n--- Concrete Slab Estimation ---")
    l = float(input("Enter Length (m): "))
    w = float(input("Enter Width (m): "))
    t = float(input("Enter Thickness (m): "))
    
    # Call the concrete function
    project_report = estimate_concrete(l, w, t)

elif choice == "2":
    print("\n--- Brick Wall Estimation ---")
    l = float(input("Enter Length of Wall (m): "))
    h = float(input("Enter Height of Wall (m): "))
    t = float(input("Enter Thickness of Wall (m) [e.g., 0.23 for 9-inch wall]: "))
    
    # Call the brick function
    project_report = estimate_brickwork(l, h, t)

else:
    print("\n❌ Invalid choice! Please restart the cell and enter 1 or 2.")


# --- DYNAMIC BILL GENERATION ---
# Only print the bill if project_report actually has data inside it
if project_report:
    print("\n" + "="*40)
    print("       BILL OF QUANTITIES (BOQ)       ")
    print("="*40)

    # Loop through the results dynamically
    for material, quantity in project_report.items():
        if material == "Total Cost (INR)":
            continue
        print(f" {material:<20} : {quantity}")

    print("-"*40)
    print(f" ESTIMATED BUDGET TOTAL: ₹{project_report['Total Cost (INR)']}/-")
    print("="*40)
    print("\n Calculations successfully generated based on site inputs.")
```
